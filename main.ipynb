```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv('/Users/youdengsong/Desktop/Data_Mining_project/Data/new_data.csv')
```

```python
df = pd.DataFrame(data)
# Group the data by 'Region Name' and calculate the average for age columns
grouped_df = df.groupby('Region Name')[['Age00', 'Age01_04', 'Age10_14', 'Age15_19', 'Age05_09', '[50-54]', '[55-59]', '[60-64]', '[65-69]', '[70-74]', '[75-79]', '[80-84]', '[85+]', 'Age_all']].mean().reset_index()

median_df = df.groupby('Region Name')[['Age00', 'Age01_04', 'Age10_14', 'Age15_19', 'Age05_09', '[50-54]', '[55-59]', '[60-64]', '[65-69]', '[70-74]', '[75-79]', '[80-84]', '[85+]', 'Age_all']].median().reset_index()



```

```python
# Create a figure with six subplots arranged in a 2x3 grid
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Average Values for Age Groups in Different Regions', fontsize=16)

# Extract unique region names from the 'Region Name' column
region_names = grouped_df['Region Name'].unique()

# Loop through each region and create a subplot for it
for i, region_name in enumerate(region_names):
    region_data = grouped_df[grouped_df['Region Name'] == region_name]
    
    # Extract age group columns (x-axis)
    age_groups = region_data.columns[1:-1]
    
    # Extract average values (y-axis)
    values = region_data.iloc[:, 1:-1].values.flatten()
    
    # Determine the subplot's position in the grid
    row = i // 3
    col = i % 3
    
    # Create a bar plot in the current subplot
    axes[row, col].bar(age_groups, values)
    axes[row, col].set_xlabel('Age Groups')
    axes[row, col].set_ylabel('Average Values')
    axes[row, col].set_title(f'{region_name}')
    axes[row, col].tick_params(axis='x', rotation=45)

# Adjust the layout and spacing
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Assuming you have your original DataFrame named 'grouped_df'
# Select the desired columns and calculate their mean for each region
selected_columns = ['Region Name', 'Country Name', 'Age00', 'Age01_04', 'Age10_14', 'Age15_19', 'Age05_09',
                    '[50-54]', '[55-59]', '[60-64]', '[65-69]', '[70-74]', '[75-79]',
                    '[80-84]', '[85+]', 'Age_all']

# Create a new DataFrame with the selected columns and their average values
new_df = df[selected_columns]

# Now, 'new_df' contains the desired columns and their average values for each region
print(new_df)
```

```python
# Step 2: Prepare your data (select the relevant columns)
data = new_df[['Age00', 'Age01_04', 'Age10_14', 'Age15_19', 'Age05_09',
               '[50-54]', '[55-59]', '[60-64]', '[65-69]', '[70-74]',
               '[75-79]', '[80-84]', '[85+]', 'Age_all']]

# Step 3: Standardize the data (optional but recommended)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 4: Perform K-means clustering
kmeans = KMeans(n_clusters=5)  # You can change the number of clusters as needed
new_df['Cluster'] = kmeans.fit_predict(scaled_data)

# Step 5: Visualize the results (optional)
plt.figure(figsize=(10, 6))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=new_df['Cluster'], cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()

# You can also examine the clusters in your DataFrame
print(new_df)
```

```python
# Calculate the number of instances in each cluster
cluster_counts = new_df['Cluster'].value_counts()
total_number = cluster_counts.sum()
# Print the number of instances in each cluster
print("Number of Instances in Each Cluster:")
print(f"{cluster_counts/total_number * 100}")
print(cluster_counts)
```

```python
# Calculate the Silhouette Score
silhouette_avg = silhouette_score(scaled_data, new_df['Cluster'])
print(f"Silhouette Score: {silhouette_avg:.2f}")
```

```python
# Find the cluster with the least number of instances
cluster_with_least_instances = cluster_counts.idxmin()

# Filter the DataFrame to include only the rows in the cluster with the least instances
countries_in_least_cluster = new_df[new_df['Cluster'] == cluster_with_least_instances]

# Print the countries in the cluster with the least instances
print("Countries in the Cluster with the Least Instances:")
print(countries_in_least_cluster[['Region Name', 'Country Name']])
```

```python
# Get the unique country names in the cluster with the least instances
unique_countries_in_least_cluster = countries_in_least_cluster['Country Name'].unique()

# Print the unique country names
print("Unique Country Names in the Cluster with the Least Instances:")
for country_name in unique_countries_in_least_cluster:
    print(country_name)
```
